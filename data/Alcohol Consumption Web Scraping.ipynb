```r
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

browser = webdriver.Chrome("/Users/jacob/Downloads/chromedriver")

def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)

statesdf = pd.DataFrame()

for i in ["https://pubs.niaaa.nih.gov/publications/surveillance110/tab4-1_16.htm", 
          "https://pubs.niaaa.nih.gov/publications/surveillance110/tab4-2_16.htm", 
          "https://pubs.niaaa.nih.gov/publications/surveillance110/tab4-3_16.htm", 
          "https://pubs.niaaa.nih.gov/publications/surveillance110/tab4-4_16.htm", 
          "https://pubs.niaaa.nih.gov/publications/surveillance110/tab4-5_16.htm", 
          "https://pubs.niaaa.nih.gov/publications/surveillance110/tab4-6_16.htm"]:
    browser.get(i)
    time.sleep(10)
    df = pd.read_html(str(BeautifulSoup(browser.page_source, "xml").find('table')), header = 0)[0]
    
    x = []
    
    for i, j in enumerate(df['State  or other  geographic area']):
        if hasNumbers(j): continue
        x.append([i,j])
    
    fulldf = pd.DataFrame()
    for i in range(len(x)):
        print(i)
        if i == 0: continue
        if i != len(x) - 1:
            newdf = df.iloc[x[i][0]:x[i+1][0], :]
        else: 
            newdf = df.iloc[x[i][0]:, :]
        newdf['State'] = x[i][1]
        fulldf = fulldf.append(newdf)
    
    statesdf = statesdf.append(fulldf)
    
```
